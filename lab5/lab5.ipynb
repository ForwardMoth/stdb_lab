### пункт a (Проверка нормальности распределения дохода)

```python
import scipy.stats as sts
import numpy as np
```

```python
df2 = df1.copy()
```

```python
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df2.Income)
print(f"Статистика критерия {z}, p_value {p}")
```

#### Гипотеза H0 о нормальном распределении отклоняется p < 0.05

```python
df2['log_income'] = np.log(df2.Income + 1)
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df2.log_income)
print(f"Статистика критерия {z}, p_value {p}")
```

#### Гипотеза H0 о нормальном распределении отклоняется p < 0.05

```python
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df2.log_income[:100])
print(f"Статистика критерия {z}, p_value {p}")
```

#### Гипотеза H0 о нормальном распределении принимается p > 0.05
