# 3 Задание

```python
df1.dropna(inplace=True)
df1
```

```python
df1 = df1[(df1['Gender'] == 'male') | (df1['Gender'] == 'female')]
df1
```

# 4, 5 Задания

### пункт a

```python
import scipy.stats as sts
import numpy as np
```

```python
df2 = df1.copy()
```

```python
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df2.Income)
print(f"Статистика критерия {z}, p_value {p}")
```

```python
df2['log_income'] = np.log(df2.Income + 1)
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df2.log_income)
print(f"Статистика критерия {z}, p_value {p}")
```

### пункт b

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols 
```

```python
model =  ols("Income ~ SchoolDegree", df2).fit()
model.summary()
```

```python
df3 = df2[:100]
```

```python
print(f"Проверка гипотезы на нормальность распределения тестом Шапиро Уилка для уровня дохода:")
z, p = sts.shapiro(df3.log_income)
print(f"Статистика критерия {z}, p_value {p}")
```

```python
model =  ols("Income ~ SchoolDegree", df3).fit()
model.summary()
```

```python
anova = sm.stats.anova_lm(model, typ=1)
anova
```

#### Вывод: Доход от уровня образования не зависит
