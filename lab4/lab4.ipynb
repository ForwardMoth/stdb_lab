```python
from scipy.stats import chi2_contingency, fisher_exact
import rpy2.robjects.numpy2ri
from rpy2.robjects.packages import importr
import numpy as np
```

### Функция, которая проверяет матрицу и определяет какой тест нужно применить

```python
def check_matrix(matrix):
    TEST = ['Хи-квадрат', 'Хи-квадрат с поправкой Йейтса', 'Точечный анализ Фишера', 'Тест Фримана Холтона']
    result = TEST[0]
    if matrix.size == 4:
        c = 0 
        for row in matrix:
            for i in row:
                if i < 5:
                    result = TEST[2]
                    break 
                elif i <= 9 and i >= 5:
                    result = TEST[1]
                elif i > 10:
                    c += 1
        
        if result != TEST[1] and result != TEST[2] and c / matrix.size > 0.8:
            result = TEST[0]
    else:
        c = 0 
        for row in matrix:
            for i in row:
                if i < 5:
                    c += 1 
        
        if c / matrix.size >= 0.8:
            result = TEST[3]
            
    print("Используем тест:", result)
    return result 
```

### Функция, которая для матрицы осуществляет нужный тест

```python
def matrix_test(matrix, name_test):
    TEST = {'Хи-квадрат': 0, 'Хи-квадрат с поправкой Йейтса': 1, 'Точечный анализ Фишера': 2, 'Тест Фримана Холтона': 3}
    if TEST[name_test] == 0:
        print(chi2_contingency(matrix, correction = False))
    elif TEST[name_test] == 1:
        print(chi2_contingency(matrix, correction = True))
    elif TEST[name_test] == 2:
        print(fisher_exact(matrix))
    else:
        rpy2.robjects.numpy2ri.activate()
        stats = importr('stats')
        m = np.array(result.values)
        res = stats.fisher_test(m, simulate_p_value = True)
        print(res)
```

### Gender, SchoolDegree

#### Таблица сопряжённости

```python
table1 = pd.crosstab(df1.Gender, df1.SchoolDegree, margins=True)
table1
```

#### Таблица ожидаемых значений

```python
result = pd.crosstab(df1.Gender, df1.SchoolDegree, margins=False)
excepted = scipy.stats.contingency.expected_freq(result)
excepted
```

#### Опеределение теста 

```python
name_test = check_matrix(excepted)
```

#### Проверка матрицы на выбранном тесте

```python
matrix_test(result, name_test)
```

### Результат: 
##### гипотеза H0 о незначимости различий между переменными отклоняется p-value < 0.05 => можно говорить о связи между полом и степенью образования
